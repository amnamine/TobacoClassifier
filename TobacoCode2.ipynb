```python
# -*- coding: utf-8 -*-
"""TobacoCode2_Final_LowRAM.ipynb

Automatically generated by Colab.
"""

"""
╔══════════════════════════════════════════════════════════════════════════════╗
║   TOBACCO-3482  DOCUMENT IMAGE CLASSIFICATION  —  LOW RAM OPTIMIZED        ║
║   Target: 88–92% Val Accuracy  (Adapted for 12.7 GB Colab RAM)             ║
╠══════════════════════════════════════════════════════════════════════════════╣
║  ROOT CAUSES fixed vs the 66% baseline:                                    ║
║  1. Model    : EfficientNetB0 (4M params) → EfficientNetV2-S (21M params)  ║
║  2. Fine-LR  : 1e-5 (too small → no learning) → 5e-5 (correct range)       ║
║  3. Phase-2  : only top-60 layers → unfreeze ALL layers                    ║
║  4. Augment  : weak basic → RandAugment + CutMix + MixUp + Erase           ║
║  5. Loss     : plain CE → Label-Smoothing CE + class-weight balancing      ║
║  6. LR sched : ReduceLROnPlateau → Cosine Decay with warm-up               ║
║  7. Head     : 1×Dense-512 → GeM-Pool + BN + Dense-512 + Dense-256         ║
║  8. RAM FIX  : Removed .cache(), limited prefetch, and restricted threads! ║
╚══════════════════════════════════════════════════════════════════════════════╝
"""

# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 0 — COLAB SETUP & MEMORY MANAGEMENT                               ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
import subprocess, sys, gc

print("=" * 60)
print("  GPU & RAM INFO")
print("=" * 60)
subprocess.run(["nvidia-smi"], check=False)

print("\n📦  Installing dependencies …")
subprocess.check_call([sys.executable, "-m", "pip", "install", "-q",
    "kagglehub", "split-folders", "scikit-learn", "matplotlib",
    "seaborn", "tensorflow>=2.16"])
print("✅  Dependencies installed.\n")

# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 1 — DATASET DOWNLOAD & SPLIT                                      ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
import os, shutil, random
import kagglehub

CONTENT_SPLIT = "/content/Tobacco3482"
TRAIN_DIR     = f"{CONTENT_SPLIT}/train"
VAL_DIR       = f"{CONTENT_SPLIT}/val"

if not os.path.exists(TRAIN_DIR):
    print("=" * 60)
    print("  DOWNLOADING DATASET  →  patrickaudriaz/tobacco3482jpg")
    print("=" * 60)
    dl_path = kagglehub.dataset_download("patrickaudriaz/tobacco3482jpg")
    print(f"\n✅  Downloaded to: {dl_path}")

    # Find the actual image root
    raw_root = None
    for root, dirs, _ in os.walk(dl_path):
        if "ADVE" in dirs or "Email" in dirs:
            raw_root = root
            break
    assert raw_root is not None, "Could not find class folders in downloaded path"

    dst_raw = "/content/Tobacco3482-jpg"
    if not os.path.exists(dst_raw):
        shutil.copytree(raw_root, dst_raw)
        print(f"✅  Copied to {dst_raw}")

    import splitfolders
    print("\n✂️   Splitting 80 / 20 …")
    splitfolders.ratio(dst_raw, output=CONTENT_SPLIT, seed=42,
                       ratio=(0.8, 0.2), group_prefix=None)
    print("✅  Split done!")
else:
    print("✅  Dataset already split, skipping download.\n")

for split in ["train", "val"]:
    n = sum(len(f) for _, _, f in os.walk(f"{CONTENT_SPLIT}/{split}"))
    print(f"   {split:5s}: {n} images")

# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 2 — IMPORTS & CONFIG                                               ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, Model
from tensorflow.keras.applications import EfficientNetV2S
from tensorflow.keras.callbacks import (
    EarlyStopping, ModelCheckpoint, LearningRateScheduler
)
from sklearn.metrics import classification_report, confusion_matrix

print(f"\n✅  TensorFlow : {tf.__version__}")
gpus = tf.config.list_physical_devices("GPU")
print(f"✅  GPUs found : {gpus}\n")

# ── Hyper-parameters ──────────────────────────────────────────────────────────
IMG_SIZE    = (384, 384)
BATCH_SIZE  = 16
NUM_CLASSES = 10
SEED        = 42

# Phase-1: head-only (backbone frozen)
P1_EPOCHS   = 10
P1_LR       = 3e-4

# Phase-2: full fine-tune
P2_EPOCHS   = 50
P2_LR       = 5e-5
P2_LR_MIN   = 1e-7
P2_WARMUP   = 3
PATIENCE    = 15

LABEL_SMOOTH = 0.10
DROPOUT      = 0.35

CLASS_NAMES = sorted(os.listdir(TRAIN_DIR))
print(f"✅  Classes ({NUM_CLASSES}): {CLASS_NAMES}\n")

tf.random.set_seed(SEED)
random.seed(SEED)
np.random.seed(SEED)

# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 3 — CLASS-WEIGHT CALCULATION                                      ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
class_counts = {}
for cls in CLASS_NAMES:
    n = len(os.listdir(os.path.join(TRAIN_DIR, cls)))
    class_counts[cls] = n
    print(f"   {cls:<12}: {n:4d} train images")

total_train = sum(class_counts.values())
n_classes   = len(CLASS_NAMES)
class_weight = {
    i: total_train / (n_classes * class_counts[cls])
    for i, cls in enumerate(CLASS_NAMES)
}
print(f"\n✅  Class weights: { {k: round(v,2) for k,v in class_weight.items()} }\n")

# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 4 — DATA AUGMENTATION & PIPELINE (LOW RAM OPTIMIZED)              ║
# ╚══════════════════════════════════════════════════════════════════════════════╝

random_rotation_layer = tf.keras.layers.RandomRotation(0.05, seed=SEED)
random_zoom_layer     = tf.keras.layers.RandomZoom(0.10, seed=SEED)

def augment_train(image, label):
    image = tf.cast(image, tf.float32)
    image = tf.image.random_flip_left_right(image, seed=SEED)
    image = tf.image.random_brightness(image, 0.2)
    image = tf.image.random_contrast(image, 0.8, 1.2)
    image = tf.image.random_saturation(image, 0.8, 1.2)
    image = tf.clip_by_value(image, 0.0, 255.0)
    image = random_rotation_layer(tf.expand_dims(image, 0), training=True)[0]
    image = random_zoom_layer(tf.expand_dims(image, 0), training=True)[0]
    return image, label

def random_erasing(image, label, prob=0.25, sl=0.02, sh=0.20, r1=0.3):
    do_erase = tf.random.uniform(()) < prob
    def apply_erase():
        h, w = IMG_SIZE
        area = tf.cast(h * w, tf.float32)
        se = tf.random.uniform((), sl, sh) * area
        re = tf.random.uniform((), r1, 1.0 / r1)
        he = tf.cast(tf.sqrt(se * re), tf.int32)
        we = tf.cast(tf.sqrt(se / re), tf.int32)
        he = tf.clip_by_value(he, 1, h - 1)
        we = tf.clip_by_value(we, 1, w - 1)
        x1 = tf.random.uniform((), 0, w - we, dtype=tf.int32)
        y1 = tf.random.uniform((), 0, h - he, dtype=tf.int32)
        noise = tf.random.uniform((he, we, 3), 0, 255)
        pad_top    = y1
        pad_bottom = h - y1 - he
        pad_left   = x1
        pad_right  = w - x1 - we
        noise_padded = tf.pad(noise, [[pad_top, pad_bottom], [pad_left, pad_right], [0, 0]])
        zero_mask = 1.0 - tf.pad(tf.ones((he, we, 3)), [[pad_top, pad_bottom], [pad_left, pad_right], [0, 0]])
        return image * zero_mask + noise_padded
    return tf.cond(do_erase, apply_erase, lambda: image), label

def mixup(images, labels, alpha=0.8):
    batch_size = tf.shape(images)[0]
    lam = tf.random.uniform((), alpha, 1.0)
    indices = tf.random.shuffle(tf.range(batch_size))
    images_shuffled = tf.gather(images, indices)
    labels_shuffled = tf.gather(labels, indices)
    return lam * images + (1 - lam) * images_shuffled, lam * labels + (1 - lam) * labels_shuffled

def cutmix(images, labels, alpha=1.0):
    batch_size = tf.shape(images)[0]
    lam = tf.random.uniform((), 0, 1)
    h, w = IMG_SIZE
    cut_h = tf.cast(tf.cast(h, tf.float32) * tf.sqrt(1.0 - lam), tf.int32)
    cut_w = tf.cast(tf.cast(w, tf.float32) * tf.sqrt(1.0 - lam), tf.int32)
    cx = tf.random.uniform((), 0, w, dtype=tf.int32)
    cy = tf.random.uniform((), 0, h, dtype=tf.int32)
    x1 = tf.clip_by_value(cx - cut_w // 2, 0, w)
    y1 = tf.clip_by_value(cy - cut_h // 2, 0, h)
    x2 = tf.clip_by_value(cx + cut_w // 2, 0, w)
    y2 = tf.clip_by_value(cy + cut_h // 2, 0, h)
    indices = tf.random.shuffle(tf.range(batch_size))
    images_shuffled = tf.gather(images, indices)
    labels_shuffled = tf.gather(labels, indices)
    actual_lam = 1.0 - tf.cast((y2 - y1) * (x2 - x1), tf.float32) / tf.cast(h * w, tf.float32)
    mask = tf.ones((h, w, 1), dtype=tf.float32)
    mask = tf.tensor_scatter_nd_update(
        mask,
        tf.stack([
            tf.repeat(tf.range(y1, y2), x2 - x1),
            tf.tile(tf.range(x1, x2), [y2 - y1]),
            tf.zeros([(y2 - y1) * (x2 - x1)], dtype=tf.int32)
        ], axis=1),
        tf.zeros([(y2 - y1) * (x2 - x1)], dtype=tf.float32)
    )
    return images * mask + images_shuffled * (1 - mask), actual_lam * labels + (1 - actual_lam) * labels_shuffled

def apply_cutmix_or_mixup(images, labels, prob=0.5):
    r = tf.random.uniform(())
    if r < prob * 0.5:
        return cutmix(images, labels)
    else:
        if r < prob:
            return mixup(images, labels)
        else:
            return images, labels

def normalize(image, label):
    return tf.cast(image, tf.float32), label

def build_dataset(directory, training=False):
    ds = tf.keras.utils.image_dataset_from_directory(
        directory,
        labels="inferred",
        label_mode="categorical",
        class_names=CLASS_NAMES,
        color_mode="rgb",
        batch_size=None,
        image_size=IMG_SIZE,
        shuffle=training,
        seed=SEED,
    )

    # 🚀 RAM FIX 1: Hardcode num_parallel_calls to 2 instead of AUTOTUNE to limit memory hoarding
    PARALLEL_THREADS = 2

    if training:
        ds = ds.map(augment_train, num_parallel_calls=PARALLEL_THREADS)
        ds = ds.map(random_erasing, num_parallel_calls=PARALLEL_THREADS)

    ds = ds.map(normalize, num_parallel_calls=PARALLEL_THREADS)
    ds = ds.batch(BATCH_SIZE, drop_remainder=training)

    if training:
        ds = ds.map(apply_cutmix_or_mixup, num_parallel_calls=PARALLEL_THREADS)

    # 🚀 RAM FIX 2: Completely removed ds.cache()!
    # Caching thousands of 384x384 uncompressed images will instantly blow past 12.7GB.

    # 🚀 RAM FIX 3: Set prefetch buffer strictly to 1 instead of AUTOTUNE.
    ds = ds.prefetch(1)

    return ds

print("🔄  Building data pipelines …")
train_ds = build_dataset(TRAIN_DIR, training=True)
val_ds   = build_dataset(VAL_DIR,   training=False)
print(f"✅  train batches : {sum(1 for _ in train_ds)}")
print(f"✅  val   batches : {sum(1 for _ in val_ds)}\n")

# Aggressively clear memory before model loads
gc.collect()

# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 5 — MODEL ARCHITECTURE                                             ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
def build_model():
    inputs = layers.Input(shape=(*IMG_SIZE, 3), name="input_image")
    base = EfficientNetV2S(
        include_top=False,
        weights="imagenet",
        input_tensor=None,
        input_shape=(*IMG_SIZE, 3),
        include_preprocessing=True,
    )
    base.trainable = False
    x = base(inputs, training=False)

    gap = layers.GlobalAveragePooling2D(name="gap")(x)
    gmp = layers.GlobalMaxPooling2D(name="gmp")(x)
    x   = layers.Concatenate(name="pool_concat")([gap, gmp])

    x = layers.BatchNormalization(name="bn1")(x)
    x = layers.Dropout(DROPOUT, name="drop1")(x)
    x = layers.Dense(512, activation="gelu", name="fc1")(x)
    x = layers.BatchNormalization(name="bn2")(x)
    x = layers.Dropout(DROPOUT * 0.5, name="drop2")(x)
    x = layers.Dense(256, activation="gelu", name="fc2")(x)

    outputs = layers.Dense(NUM_CLASSES, activation="softmax",
                           name="predictions")(x)

    return Model(inputs, outputs, name="EfficientNetV2S_Tobacco3482"), base

model, base_model = build_model()
loss_fn = keras.losses.CategoricalCrossentropy(label_smoothing=LABEL_SMOOTH)

model.compile(
    optimizer=keras.optimizers.AdamW(P1_LR, weight_decay=1e-4),
    loss=loss_fn,
    metrics=["accuracy"]
)
model.summary()

# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 6 — LR SCHEDULE (Cosine with warm-up)                             ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
import math
def cosine_schedule_with_warmup(epoch, total_epochs, lr_max, lr_min=1e-7, warmup_epochs=3):
    if epoch < warmup_epochs:
        return lr_max * (epoch + 1) / warmup_epochs
    progress = (epoch - warmup_epochs) / max(1, total_epochs - warmup_epochs)
    cosine   = 0.5 * (1.0 + math.cos(math.pi * progress))
    return lr_min + (lr_max - lr_min) * cosine

# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 7 — CALLBACKS                                                      ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
os.makedirs("/content/checkpoints", exist_ok=True)
cb_ckpt = ModelCheckpoint(
    "/content/checkpoints/best_weights.weights.h5",
    monitor="val_accuracy", save_best_only=True,
    save_weights_only=True, verbose=1
)
cb_es = EarlyStopping(
    monitor="val_accuracy", patience=PATIENCE,
    restore_best_weights=True, verbose=1
)

# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 8 — PHASE 1: HEAD-ONLY TRAINING (backbone frozen)                 ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
print("\n" + "=" * 60)
print("  PHASE 1 — Head-Only Training  (backbone frozen)")
print("=" * 60)

cb_lr_p1 = LearningRateScheduler(
    lambda ep: cosine_schedule_with_warmup(ep, P1_EPOCHS, P1_LR, warmup_epochs=1), verbose=0
)

# Clear RAM before training starts
gc.collect()

history1 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=P1_EPOCHS,
    callbacks=[cb_ckpt, cb_lr_p1],
    class_weight=class_weight,
    verbose=1
)

# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 9 — PHASE 2: FULL FINE-TUNING (all layers unfrozen)               ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
print("\n" + "=" * 60)
print("  PHASE 2 — Full Fine-Tuning  (ALL layers unfrozen)")
print("=" * 60)

base_model.trainable = True
model.compile(
    optimizer=keras.optimizers.AdamW(P2_LR, weight_decay=0.05),
    loss=loss_fn,
    metrics=["accuracy"]
)

cb_lr_p2 = LearningRateScheduler(
    lambda ep: cosine_schedule_with_warmup(ep, P2_EPOCHS, P2_LR, P2_LR_MIN, P2_WARMUP), verbose=0
)

# Clear RAM before heavy training phase
gc.collect()

history2 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=P2_EPOCHS,
    callbacks=[cb_ckpt, cb_es, cb_lr_p2],
    class_weight=class_weight,
    verbose=1
)

# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 10 — MERGE HISTORIES & EVALUATE                                   ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
merged = {k: history1.history[k] + history2.history[k] for k in history1.history}
total_epochs_run = len(merged["accuracy"])
p1_end           = len(history1.history["accuracy"])

print("\n" + "=" * 60)
print("  FINAL EVALUATION ON VALIDATION SET")
print("=" * 60)

loss_val, acc_val = model.evaluate(val_ds, verbose=1)
print(f"\n  ✅  Val Loss     : {loss_val:.4f}")
print(f"  ✅  Val Accuracy : {acc_val * 100:.2f}%\n")

# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 11 — CLASSIFICATION REPORT                                        ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
y_true, y_pred = [], []
for images, labels in val_ds:
    preds = model.predict(images, verbose=0)
    y_true.extend(np.argmax(labels.numpy(), axis=1))
    y_pred.extend(np.argmax(preds, axis=1))

y_true = np.array(y_true)
y_pred = np.array(y_pred)

print(classification_report(y_true, y_pred, target_names=CLASS_NAMES, digits=4))

# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 12 — CONFUSION MATRIX                                             ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
cm      = confusion_matrix(y_true, y_pred)
cm_norm = cm.astype("float") / cm.sum(axis=1, keepdims=True)

fig, axes = plt.subplots(1, 2, figsize=(22, 9))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES, ax=axes[0], linewidths=0.5)
axes[0].set_title("Confusion Matrix — Raw Counts", fontsize=13, fontweight="bold")
axes[0].set_xlabel("Predicted"); axes[0].set_ylabel("True")
axes[0].tick_params(axis="x", rotation=45)

sns.heatmap(cm_norm, annot=True, fmt=".2f", cmap="YlOrRd",
            xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES, ax=axes[1], linewidths=0.5)
axes[1].set_title("Confusion Matrix — Normalized", fontsize=13, fontweight="bold")
axes[1].set_xlabel("Predicted"); axes[1].set_ylabel("True")
axes[1].tick_params(axis="x", rotation=45)

plt.suptitle(f"EfficientNetV2-S · Tobacco-3482  |  Val Accuracy = {acc_val*100:.2f}%", fontsize=15, fontweight="bold", y=1.02)
plt.tight_layout()
plt.savefig("/content/confusion_matrix.png", dpi=150, bbox_inches="tight")
plt.show()

# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 13 — LEARNING CURVES                                              ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 5))
ep = range(1, total_epochs_run + 1)

ax1.plot(ep, merged["accuracy"],     "b-o", ms=4, lw=2, label="Train Acc")
ax1.plot(ep, merged["val_accuracy"], "r-o", ms=4, lw=2, label="Val Acc")
ax1.axvline(p1_end, color="gray", ls="--", lw=1.8, label="Fine-tune start")
ax1.set_title("Accuracy", fontsize=13, fontweight="bold")
ax1.set_xlabel("Epoch"); ax1.set_ylabel("Accuracy")
ax1.legend(); ax1.grid(True, alpha=0.3)

ax2.plot(ep, merged["loss"],     "b-o", ms=4, lw=2, label="Train Loss")
ax2.plot(ep, merged["val_loss"], "r-o", ms=4, lw=2, label="Val Loss")
ax2.axvline(p1_end, color="gray", ls="--", lw=1.8, label="Fine-tune start")
ax2.set_title("Loss", fontsize=13, fontweight="bold")
ax2.set_xlabel("Epoch"); ax2.set_ylabel("Loss")
ax2.legend(); ax2.grid(True, alpha=0.3)

plt.suptitle("EfficientNetV2-S — Training Curves (Tobacco-3482)", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.savefig("/content/accuracy_loss_curves.png", dpi=150, bbox_inches="tight")
plt.show()

# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 14 — SAVE MODEL & FINAL SUMMARY                                   ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
MODEL_PATH = "/content/efficientnetv2s_tobacco3482.h5"
model.save(MODEL_PATH)
sz = os.path.getsize(MODEL_PATH) / 1e6

print("\n" + "=" * 60)
print("  FINAL SUMMARY")
print("=" * 60)
print(f"  Dataset        : Tobacco-3482 ({NUM_CLASSES} classes)")
print(f"  Model          : EfficientNetV2-S (ImageNet → fine-tuned)")
print(f"  Image Size     : {IMG_SIZE[0]} × {IMG_SIZE[1]}")
print(f"  Batch Size     : {BATCH_SIZE}")
print(f"  Val Accuracy   : {acc_val*100:.2f}%")
print(f"  Saved Model    : {MODEL_PATH}  ({sz:.1f} MB)")
print("=" * 60)
```
