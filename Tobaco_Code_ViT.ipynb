```python
# -*- coding: utf-8 -*-
"""
╔══════════════════════════════════════════════════════════════════════════════╗
║   TOBACCO-3482  —  VISION TRANSFORMER (ViT)  CLASSIFICATION  [FIXED v4]    ║
║   Hardware : Free Colab — 12.7 GB RAM  +  15 GB T4 GPU                     ║
╠══════════════════════════════════════════════════════════════════════════════╣
║  FIXES vs previous version:                                                  ║
║  1. tqdm progress bars on EVERY batch (train + val)                         ║
║  2. Google Drive mounted  → all weights saved to MyDrive/TOBACCO/           ║
║  3. Per-epoch checkpoint saved (epoch_001.pt … epoch_N.pt)                  ║
║  4. grad_clip added (prevented exploding gradients → training death)        ║
║  5. HF forward() fixed: pass pixel_values=, not positional                  ║
║  6. AMP scaler reset between phases (stale scale caused NaN loss)           ║
║  7. torch.load() uses weights_only=True (deprecation fix)                   ║
║  8. tqdm uses dynamic_ncols=True + unit='batch' so Colab renders cleanly    ║
╚══════════════════════════════════════════════════════════════════════════════╝
"""

# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 0 — INSTALL                                                        ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
import subprocess, sys

print("=" * 65)
print("  TOBACCO-3482  |  ViT Classification  |  NO CNN")
print("=" * 65)

subprocess.run(["nvidia-smi"], check=False)

print("\n📦 Installing packages …")
subprocess.check_call([
    sys.executable, "-m", "pip", "install", "-q",
    "kagglehub",
    "split-folders",
    "scikit-learn",
    "matplotlib",
    "seaborn",
    "Pillow",
    "timm>=0.9.0",
    "transformers>=4.40.0",
    "torch",
    "torchvision",
    "tqdm",
])
print("✅ All packages installed.\n")


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 1 — MOUNT GOOGLE DRIVE                                             ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
import os, shutil, random, gc, math

from google.colab import drive
print("🔗 Mounting Google Drive …")
drive.mount("/content/drive", force_remount=False)

# All weights + plots will go here — survives Colab disconnects
DRIVE_SAVE_DIR = "/content/drive/MyDrive/TOBACCO"
os.makedirs(DRIVE_SAVE_DIR, exist_ok=True)
print(f"✅ Drive mounted. Saving everything to: {DRIVE_SAVE_DIR}\n")


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 2 — IMPORTS & DEVICE                                               ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms, datasets

from sklearn.metrics import classification_report, confusion_matrix
from tqdm.auto import tqdm           # ← auto selects notebook / terminal bar

import kagglehub
import splitfolders

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"✅ PyTorch : {torch.__version__}")
print(f"✅ Device  : {DEVICE}")
if torch.cuda.is_available():
    print(f"✅ GPU     : {torch.cuda.get_device_name(0)}")
    print(f"✅ VRAM    : {torch.cuda.get_device_properties(0).total_memory / 1e9:.1f} GB\n")


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 3 — CONFIG                                                         ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
SEED        = 42
IMG_SIZE    = 224          # ViT-B/16 standard
BATCH_SIZE  = 16           # Safe for 15 GB T4
NUM_CLASSES = 10
NUM_WORKERS = 2

P1_EPOCHS   = 5            # head warm-up   (backbone frozen)
P1_LR       = 1e-3

P2_EPOCHS   = 10           # full fine-tune (all layers)
P2_LR       = 2e-5
PATIENCE    = 5            # early-stop patience for Phase 2
GRAD_CLIP   = 1.0          # gradient clipping — prevents training collapse

CONTENT_SPLIT = "/content/Tobacco3482"
TRAIN_DIR     = f"{CONTENT_SPLIT}/train"
VAL_DIR       = f"{CONTENT_SPLIT}/val"

random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)

print("✅ Config set.")


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 4 — DATASET DOWNLOAD & SPLIT                                       ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
if not os.path.exists(TRAIN_DIR):
    print("=" * 65)
    print("  DOWNLOADING  →  patrickaudriaz/tobacco3482jpg")
    print("=" * 65)
    dl_path = kagglehub.dataset_download("patrickaudriaz/tobacco3482jpg")
    print(f"✅ Downloaded to: {dl_path}")

    raw_root = None
    for root, dirs, _ in os.walk(dl_path):
        if "ADVE" in dirs or "Email" in dirs:
            raw_root = root
            break
    assert raw_root, "❌ Cannot find class folders in download"

    dst_raw = "/content/Tobacco3482-jpg"
    if not os.path.exists(dst_raw):
        shutil.copytree(raw_root, dst_raw)

    print("✂️  Splitting 80/20 …")
    splitfolders.ratio(dst_raw, output=CONTENT_SPLIT, seed=SEED,
                       ratio=(0.8, 0.2), group_prefix=None)
    print("✅ Split done.")
else:
    print("✅ Dataset already split — skipping download.\n")

CLASS_NAMES = sorted(os.listdir(TRAIN_DIR))
print(f"\n✅ Classes ({NUM_CLASSES}): {CLASS_NAMES}")
for split in ["train", "val"]:
    n = sum(len(f) for _, _, f in os.walk(f"{CONTENT_SPLIT}/{split}"))
    print(f"   {split}: {n} images")


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 5 — CLASS WEIGHTS                                                  ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
print("\n📊 Class distribution:")
counts = []
for cls in CLASS_NAMES:
    n = len(os.listdir(os.path.join(TRAIN_DIR, cls)))
    counts.append(n)
    print(f"   {cls:<12}: {n}")

total_imgs   = sum(counts)
class_weights = torch.tensor(
    [total_imgs / (NUM_CLASSES * c) for c in counts], dtype=torch.float32
).to(DEVICE)
print(f"\n✅ Class weights: {[round(w.item(), 2) for w in class_weights]}\n")


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 6 — DATA PIPELINE                                                  ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
train_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomApply([transforms.ColorJitter(0.2, 0.2, 0.2, 0.05)], p=0.5),
    transforms.RandomApply([transforms.RandomRotation(10)], p=0.3),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5]),
])

val_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5]),
])

train_dataset = datasets.ImageFolder(TRAIN_DIR, transform=train_transform)
val_dataset   = datasets.ImageFolder(VAL_DIR,   transform=val_transform)

assert train_dataset.classes == CLASS_NAMES, \
    f"Class mismatch: {train_dataset.classes} vs {CLASS_NAMES}"

train_loader = DataLoader(
    train_dataset, batch_size=BATCH_SIZE, shuffle=True,
    num_workers=NUM_WORKERS, pin_memory=True, drop_last=True,
    persistent_workers=True,
)
val_loader = DataLoader(
    val_dataset, batch_size=BATCH_SIZE, shuffle=False,
    num_workers=NUM_WORKERS, pin_memory=True,
    persistent_workers=True,
)

print(f"✅ Train batches : {len(train_loader)}")
print(f"✅ Val   batches : {len(val_loader)}\n")


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 7 — MODEL LOADING (4-level redundancy, still NO CNN)               ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
print("=" * 65)
print("  LOADING ViT BACKBONE (4-level redundancy)")
print("=" * 65)

backbone      = None
backbone_name = None
backbone_type = None   # "torchvision" | "timm" | "hf"

# ── Option 1: torchvision built-in ViT-B/16 ──────────────────────────────────
if backbone is None:
    try:
        import torchvision.models as tv_models
        print("\n🔄 [Option 1] torchvision ViT-B/16 …")
        weights  = tv_models.ViT_B_16_Weights.IMAGENET1K_V1
        backbone = tv_models.vit_b_16(weights=weights)
        in_feat  = backbone.heads.head.in_features
        backbone.heads.head = nn.Linear(in_feat, NUM_CLASSES)
        backbone_name = "torchvision ViT-B/16"
        backbone_type = "torchvision"
        print(f"✅ [Option 1] Loaded: {backbone_name}")
    except Exception as e:
        print(f"⚠️  [Option 1] Failed: {e}")
        backbone = None

# ── Option 2: timm ViT-B/16 ──────────────────────────────────────────────────
if backbone is None:
    try:
        import timm
        print("\n🔄 [Option 2] timm vit_base_patch16_224 …")
        backbone = timm.create_model(
            "vit_base_patch16_224", pretrained=True, num_classes=NUM_CLASSES
        )
        backbone_name = "timm ViT-B/16"
        backbone_type = "timm"
        print(f"✅ [Option 2] Loaded: {backbone_name}")
    except Exception as e:
        print(f"⚠️  [Option 2] Failed: {e}")
        backbone = None

# ── Option 3: HuggingFace ViTForImageClassification (PyTorch) ────────────────
if backbone is None:
    try:
        from transformers import ViTForImageClassification
        print("\n🔄 [Option 3] HuggingFace ViT-B/16 (PyTorch) …")
        backbone = ViTForImageClassification.from_pretrained(
            "google/vit-base-patch16-224-in21k",
            num_labels=NUM_CLASSES,
            id2label={i: c for i, c in enumerate(CLASS_NAMES)},
            label2id={c: i for i, c in enumerate(CLASS_NAMES)},
            ignore_mismatched_sizes=True,
        )
        backbone_name = "HuggingFace ViT-B/16"
        backbone_type = "hf"
        print(f"✅ [Option 3] Loaded: {backbone_name}")
    except Exception as e:
        print(f"⚠️  [Option 3] Failed: {e}")
        backbone = None

# ── Option 4: timm DeiT-Small (pure Transformer fallback) ────────────────────
if backbone is None:
    try:
        import timm
        print("\n🔄 [Option 4] timm deit_small_patch16_224 (fallback) …")
        backbone = timm.create_model(
            "deit_small_patch16_224", pretrained=True, num_classes=NUM_CLASSES
        )
        backbone_name = "timm DeiT-Small"
        backbone_type = "timm"
        print(f"✅ [Option 4] Loaded: {backbone_name}")
    except Exception as e:
        print(f"⚠️  [Option 4] Failed: {e}")
        backbone = None

if backbone is None:
    raise RuntimeError("❌ ALL 4 model options failed. Check internet connection.")

backbone = backbone.to(DEVICE)
total_params = sum(p.numel() for p in backbone.parameters())
print(f"\n🎯 Model  : {backbone_name}")
print(f"   Params : {total_params:,}  ({total_params/1e6:.1f}M)\n")


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 8 — MODEL UTILITIES                                                ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
def is_hf(m):
    try:
        from transformers import ViTForImageClassification
        return isinstance(m, ViTForImageClassification)
    except ImportError:
        return False


def freeze_backbone(m):
    """Freeze all parameters except the final head."""
    if is_hf(m):
        for p in m.vit.parameters():
            p.requires_grad = False
        for p in m.classifier.parameters():
            p.requires_grad = True
    elif hasattr(m, "heads"):          # torchvision ViT
        for name, p in m.named_parameters():
            p.requires_grad = ("heads" in name)
    else:                              # timm ViT / DeiT
        for name, p in m.named_parameters():
            p.requires_grad = ("head" in name)


def unfreeze_all(m):
    for p in m.parameters():
        p.requires_grad = True


def forward_pass(m, imgs):
    """Unified forward — handles all 3 model types correctly."""
    if is_hf(m):
        # FIX: must pass keyword argument pixel_values= for HuggingFace
        return m(pixel_values=imgs).logits
    else:
        return m(imgs)


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 9 — LOSS & SCALER                                                  ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
criterion = nn.CrossEntropyLoss(weight=class_weights, label_smoothing=0.10)

# NOTE: scaler is recreated before each phase to avoid stale scale from Phase 1
#       causing NaN loss at the start of Phase 2 (this was a key failure point)


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 10 — TRAIN / EVAL FUNCTIONS WITH tqdm                              ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
def train_one_epoch(model, loader, optimizer, scaler, epoch_label=""):
    """
    Trains one epoch.  Returns (avg_loss, accuracy).
    tqdm bar shows loss + acc updating every batch.
    """
    model.train()
    running_loss, correct, total = 0.0, 0, 0

    pbar = tqdm(
        loader,
        desc=f"  🏋️  {epoch_label} [train]",
        unit="batch",
        dynamic_ncols=True,
        leave=True,
    )

    for imgs, labels in pbar:
        imgs, labels = imgs.to(DEVICE, non_blocking=True), \
                       labels.to(DEVICE, non_blocking=True)

        optimizer.zero_grad(set_to_none=True)

        with torch.cuda.amp.autocast():
            logits = forward_pass(model, imgs)
            loss   = criterion(logits, labels)

        scaler.scale(loss).backward()

        # FIX: unscale before clip so clip sees real gradient magnitudes
        scaler.unscale_(optimizer)
        nn.utils.clip_grad_norm_(model.parameters(), GRAD_CLIP)

        scaler.step(optimizer)
        scaler.update()

        bs            = imgs.size(0)
        running_loss += loss.item() * bs
        correct      += (logits.argmax(1) == labels).sum().item()
        total        += bs

        # Live metrics in tqdm bar
        pbar.set_postfix(
            loss=f"{running_loss/total:.4f}",
            acc=f"{correct/total:.4f}",
        )

    pbar.close()
    return running_loss / total, correct / total


@torch.no_grad()
def evaluate(model, loader, epoch_label=""):
    """
    Evaluates one epoch.  Returns (avg_loss, accuracy).
    tqdm bar shows live loss + acc.
    """
    model.eval()
    running_loss, correct, total = 0.0, 0, 0

    pbar = tqdm(
        loader,
        desc=f"  📊  {epoch_label} [ val ]",
        unit="batch",
        dynamic_ncols=True,
        leave=True,
    )

    for imgs, labels in pbar:
        imgs, labels = imgs.to(DEVICE, non_blocking=True), \
                       labels.to(DEVICE, non_blocking=True)

        with torch.cuda.amp.autocast():
            logits = forward_pass(model, imgs)
            loss   = criterion(logits, labels)

        bs            = imgs.size(0)
        running_loss += loss.item() * bs
        correct      += (logits.argmax(1) == labels).sum().item()
        total        += bs

        pbar.set_postfix(
            loss=f"{running_loss/total:.4f}",
            acc=f"{correct/total:.4f}",
        )

    pbar.close()
    return running_loss / total, correct / total


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 11 — DRIVE SAVE HELPER                                             ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
def save_checkpoint(model, phase, epoch, val_acc, is_best=False):
    """
    Saves two files to MyDrive/TOBACCO/:
      1. epoch checkpoint  → phase1_epoch_003.pt
      2. best model copy   → phase1_best.pt  (only when is_best=True)
    """
    fname = f"phase{phase}_epoch_{epoch:03d}_acc{val_acc:.4f}.pt"
    fpath = os.path.join(DRIVE_SAVE_DIR, fname)
    torch.save(model.state_dict(), fpath)
    print(f"      💾 Saved → {fpath}")

    if is_best:
        best_path = os.path.join(DRIVE_SAVE_DIR, f"phase{phase}_best.pt")
        shutil.copy2(fpath, best_path)
        print(f"      ⭐ Best model updated → {best_path}")

    return fpath


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 12 — PHASE 1: HEAD WARM-UP (backbone frozen)                       ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
print("\n" + "=" * 65)
print(f"  PHASE 1 — Head Warm-Up  ({P1_EPOCHS} epochs, backbone FROZEN)")
print("=" * 65)

freeze_backbone(backbone)
trainable_p1 = sum(p.numel() for p in backbone.parameters() if p.requires_grad)
print(f"  Trainable params: {trainable_p1:,}\n")

optimizer_p1 = optim.Adam(
    [p for p in backbone.parameters() if p.requires_grad],
    lr=P1_LR,
)

# FIX: fresh scaler for each phase
scaler_p1 = torch.cuda.amp.GradScaler()

hist_p1        = {"train_loss": [], "train_acc": [], "val_loss": [], "val_acc": []}
best_p1_acc    = 0.0
best_p1_path   = None

for epoch in range(1, P1_EPOCHS + 1):
    ep_label = f"P1 Ep {epoch}/{P1_EPOCHS}"

    tr_loss, tr_acc = train_one_epoch(backbone, train_loader,
                                      optimizer_p1, scaler_p1, ep_label)
    vl_loss, vl_acc = evaluate(backbone, val_loader, ep_label)

    hist_p1["train_loss"].append(tr_loss)
    hist_p1["train_acc"].append(tr_acc)
    hist_p1["val_loss"].append(vl_loss)
    hist_p1["val_acc"].append(vl_acc)

    is_best = vl_acc > best_p1_acc
    if is_best:
        best_p1_acc = vl_acc

    ckpt_path = save_checkpoint(backbone, phase=1, epoch=epoch,
                                 val_acc=vl_acc, is_best=is_best)
    if is_best:
        best_p1_path = ckpt_path

    print(f"\n  ▶ Epoch {epoch:02d}/{P1_EPOCHS} summary │ "
          f"train_loss={tr_loss:.4f}  train_acc={tr_acc:.4f} │ "
          f"val_loss={vl_loss:.4f}  val_acc={vl_acc:.4f}"
          f"{'  ⭐ NEW BEST' if is_best else ''}\n")

# Restore best Phase-1 weights
backbone.load_state_dict(
    torch.load(
        os.path.join(DRIVE_SAVE_DIR, "phase1_best.pt"),
        map_location=DEVICE,
        weights_only=True,   # FIX: suppress deprecation warning
    )
)
print(f"\n✅ Phase 1 complete. Best val acc: {best_p1_acc:.4f}\n")
gc.collect()
torch.cuda.empty_cache()


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 13 — PHASE 2: FULL FINE-TUNE (all layers unfrozen)                 ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
print("=" * 65)
print(f"  PHASE 2 — Full Fine-Tune  ({P2_EPOCHS} epochs, ALL layers)")
print("=" * 65)

unfreeze_all(backbone)
trainable_p2 = sum(p.numel() for p in backbone.parameters() if p.requires_grad)
print(f"  Trainable params: {trainable_p2:,}\n")

optimizer_p2 = optim.AdamW(backbone.parameters(), lr=P2_LR, weight_decay=1e-4)
scheduler_p2 = optim.lr_scheduler.CosineAnnealingLR(
    optimizer_p2, T_max=P2_EPOCHS, eta_min=1e-7
)

# FIX: fresh scaler — old Phase-1 scaler can have a very small scale
#      which makes Phase-2 gradients vanish (silent training failure)
scaler_p2 = torch.cuda.amp.GradScaler()

hist_p2          = {"train_loss": [], "train_acc": [], "val_loss": [], "val_acc": []}
best_p2_acc      = 0.0
patience_counter = 0

for epoch in range(1, P2_EPOCHS + 1):
    ep_label = f"P2 Ep {epoch}/{P2_EPOCHS}"
    cur_lr   = optimizer_p2.param_groups[0]["lr"]

    tr_loss, tr_acc = train_one_epoch(backbone, train_loader,
                                      optimizer_p2, scaler_p2, ep_label)
    vl_loss, vl_acc = evaluate(backbone, val_loader, ep_label)
    scheduler_p2.step()

    hist_p2["train_loss"].append(tr_loss)
    hist_p2["train_acc"].append(tr_acc)
    hist_p2["val_loss"].append(vl_loss)
    hist_p2["val_acc"].append(vl_acc)

    is_best = vl_acc > best_p2_acc
    if is_best:
        best_p2_acc = vl_acc
        patience_counter = 0
    else:
        patience_counter += 1

    save_checkpoint(backbone, phase=2, epoch=epoch,
                    val_acc=vl_acc, is_best=is_best)

    print(f"\n  ▶ Epoch {epoch:02d}/{P2_EPOCHS} summary │ "
          f"lr={cur_lr:.2e} │ "
          f"train_loss={tr_loss:.4f}  train_acc={tr_acc:.4f} │ "
          f"val_loss={vl_loss:.4f}  val_acc={vl_acc:.4f}"
          f"{'  ⭐ NEW BEST' if is_best else f'  (patience {patience_counter}/{PATIENCE})'}\n")

    if patience_counter >= PATIENCE:
        print("  ⏹️  Early stopping triggered.")
        break

# Restore best Phase-2 weights
backbone.load_state_dict(
    torch.load(
        os.path.join(DRIVE_SAVE_DIR, "phase2_best.pt"),
        map_location=DEVICE,
        weights_only=True,
    )
)
print(f"\n✅ Phase 2 complete. Best val acc: {best_p2_acc:.4f}\n")
gc.collect()
torch.cuda.empty_cache()


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 14 — FINAL EVALUATION                                              ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
print("=" * 65)
print("  FINAL EVALUATION")
print("=" * 65)

final_loss, final_acc = evaluate(backbone, val_loader, "Final Eval")
print(f"\n✅ Final Val Loss : {final_loss:.4f}")
print(f"✅ Final Val Acc  : {final_acc:.4f}  ({final_acc * 100:.2f}%)\n")

# Collect all predictions
backbone.eval()
y_true, y_pred = [], []
with torch.no_grad():
    for imgs, labels in tqdm(val_loader, desc="  🔍 Collecting predictions",
                             unit="batch", dynamic_ncols=True):
        imgs = imgs.to(DEVICE, non_blocking=True)
        with torch.cuda.amp.autocast():
            logits = forward_pass(backbone, imgs)
        y_pred.extend(logits.argmax(1).cpu().tolist())
        y_true.extend(labels.tolist())

y_true = np.array(y_true)
y_pred = np.array(y_pred)

print("\n📊 Classification Report:")
print(classification_report(y_true, y_pred, target_names=CLASS_NAMES, digits=4))


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 15 — PLOTS (saved to Drive)                                        ║
# ╚══════════════════════════════════════════════════════════════════════════════╝

# ── Confusion Matrix ──────────────────────────────────────────────────────────
cm = confusion_matrix(y_true, y_pred)
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES,
            ax=axes[0], linewidths=0.4)
axes[0].set_title("Confusion Matrix — Raw Counts", fontsize=13, fontweight="bold")
axes[0].set_xlabel("Predicted"); axes[0].set_ylabel("True")
axes[0].tick_params(axis="x", rotation=45)

cm_norm = cm.astype("float") / cm.sum(axis=1, keepdims=True)
sns.heatmap(cm_norm, annot=True, fmt=".2f", cmap="YlOrRd",
            xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES,
            ax=axes[1], linewidths=0.4)
axes[1].set_title("Confusion Matrix — Normalized", fontsize=13, fontweight="bold")
axes[1].set_xlabel("Predicted"); axes[1].set_ylabel("True")
axes[1].tick_params(axis="x", rotation=45)

plt.suptitle(
    f"{backbone_name}  ·  Tobacco-3482  |  Val Acc = {final_acc*100:.2f}%",
    fontsize=14, fontweight="bold", y=1.01
)
plt.tight_layout()
cm_path = os.path.join(DRIVE_SAVE_DIR, "confusion_matrix.png")
plt.savefig(cm_path, dpi=150, bbox_inches="tight")
plt.show()
print(f"✅ Confusion matrix saved → {cm_path}")

# ── Training Curves ───────────────────────────────────────────────────────────
all_tr_acc  = hist_p1["train_acc"]  + hist_p2["train_acc"]
all_vl_acc  = hist_p1["val_acc"]    + hist_p2["val_acc"]
all_tr_loss = hist_p1["train_loss"] + hist_p2["train_loss"]
all_vl_loss = hist_p1["val_loss"]   + hist_p2["val_loss"]
p1_end      = len(hist_p1["train_acc"])
ep_axis     = range(1, len(all_tr_acc) + 1)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(ep_axis, all_tr_acc,  label="Train Acc",  color="royalblue", lw=2)
ax1.plot(ep_axis, all_vl_acc,  label="Val Acc",    color="coral",     lw=2)
ax1.axvline(p1_end, color="gray", linestyle="--", lw=1.5, label="Phase 2 start")
ax1.set_title(f"Accuracy — {backbone_name}\nTobacco-3482")
ax1.set_xlabel("Epoch"); ax1.set_ylabel("Accuracy")
ax1.legend(); ax1.grid(True, alpha=0.3)

ax2.plot(ep_axis, all_tr_loss, label="Train Loss", color="royalblue", lw=2)
ax2.plot(ep_axis, all_vl_loss, label="Val Loss",   color="coral",     lw=2)
ax2.axvline(p1_end, color="gray", linestyle="--", lw=1.5, label="Phase 2 start")
ax2.set_title(f"Loss — {backbone_name}\nTobacco-3482")
ax2.set_xlabel("Epoch"); ax2.set_ylabel("Loss")
ax2.legend(); ax2.grid(True, alpha=0.3)

plt.tight_layout()
curves_path = os.path.join(DRIVE_SAVE_DIR, "training_curves.png")
plt.savefig(curves_path, dpi=150)
plt.show()
print(f"✅ Training curves saved → {curves_path}")


# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SECTION 16 — FINAL SAVE & SUMMARY                                          ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
final_weights_path = os.path.join(DRIVE_SAVE_DIR, "vit_tobacco_FINAL.pt")
torch.save(backbone.state_dict(), final_weights_path)

print("\n" + "=" * 65)
print("  ✅  DONE — FINAL SUMMARY")
print("=" * 65)
print(f"  Model          : {backbone_name}")
print(f"  Dataset        : Tobacco-3482  ({NUM_CLASSES} classes)")
print(f"  Image size     : {IMG_SIZE}×{IMG_SIZE}")
print(f"  Batch size     : {BATCH_SIZE}")
print(f"  Phase 1 acc    : {best_p1_acc:.4f}  ({best_p1_acc*100:.2f}%)")
print(f"  Phase 2 acc    : {best_p2_acc:.4f}  ({best_p2_acc*100:.2f}%)")
print(f"  Final val acc  : {final_acc:.4f}  ({final_acc*100:.2f}%)")
print(f"  All weights in : {DRIVE_SAVE_DIR}/")
print(f"  Final weights  : {final_weights_path}")
print("=" * 65)
```
